# Data Cleaning 

## Handling missing values 

```python
df['CouponCode'].fillna('No Coupon', inplace=True)
```

```python
df.isnull().sum()
```

## checking duplicates 

```python
df[df.duplicated('OrderID')]
```

# Standardization 

## Convert date column 

```python
df['Date'] = pd.to_datetime(df['Date'])
```

## Remove invalid data 

```python
df[df['Quantity'] < 0]
df[df['TotalPrice'] < 0]
```

## Clean column names 

```python
df['Product'] = df['Product'].str.strip()
```

## Fix Business Logic 

```python
df['calculated'] = df['Quantity'] * df['UnitPrice']
df[df['calculated'] != df['TotalPrice']]
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

## Round Prices to 2 decimal digits 

```python
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)
```

```python
df.head()
```

## Clean text columns 

```python
text_cols = df.select_dtypes(include='object').columns
for col in text_cols:
    df[col] = df[col].str.strip()
df
```

## Final Validation

```python
print("\nFinal Missing Values:\n", df.isnull().sum())
```
